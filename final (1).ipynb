```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd

class FashionRecommender:
    def __init__(self, item_features, fashion_data):
        self.item_features = item_features
        self.fashion_data = fashion_data

    def recommend_items(self, input_categories, n=10):
        # Combine input categories into a single query
        query = ' '.join(input_categories)
        # Transform the query using the same vectorizer
        query_features = self.vectorizer.transform([query])
        # Calculate cosine similarity between the query and item features
        similarity_scores = cosine_similarity(query_features, self.item_features)
        # Find indices of the most similar items
        similar_items_indices = similarity_scores.argsort()[0][-n-1:-1][::-1]
        return self.fashion_data.iloc[similar_items_indices]

# Load fashion item dataset
# Assuming df is your existing DataFrame containing fashion item data
# df = pd.read_csv('styles.csv')

# Feature Engineering
df['features'] = df['subCategory'] + ' ' + df['baseColour'] + ' ' + df['articleType'] + ' ' + df['masterCategory'] + ' ' + df['usage'] + ' ' + df['gender'] + ' ' + df['productDisplayName']

# Preprocess features
vectorizer = TfidfVectorizer()
item_features = vectorizer.fit_transform(df['features'])

# Instantiate the recommender system
recommender = FashionRecommender(item_features, df)
recommender.vectorizer = vectorizer  # Save the vectorizer for later use

def find_and_display_image(image_id, image_csv_path):
    """
    Function to find and display the image based on the given image ID from an image CSV file.
    
    Parameters:
        image_id (int): The ID of the image to search for.
        image_csv_path (str): Path to the image CSV file containing ID and path columns.
    """
    try:
        # Load image CSV file
        image_df = pd.read_csv(image_csv_path)
        # Find the image path based on the provided ID
        image_path = image_df[image_df['id'] == image_id]['file_path'].values[0]
        # Open and display the image
        img = Image.open(image_path)
        plt.imshow(img)
        plt.title(f"Budget: {(image_id)/10}")
        plt.axis('off')  # Turn off axis labels
        plt.show()
    except (IndexError, FileNotFoundError):
        # Handle the case where the ID is not found or CSV file is not found
        print(f"Image path not found for ID {image_id}.")

# Example usage:
input_categories = ['Women', 'Topwear', 'Shirts', 'Tshirts', 'Pink', 'Formal']
recommendations = recommender.recommend_items(input_categories)

# Retrieve and display image for each recommended item
for index, row in recommendations.iterrows():
    image_id = row['id']  # Assuming 'id' is the column containing unique image IDs
    find_and_display_image(image_id, r"C:\Users\Vaishnavi\Desktop\hackathon copy\image_data.csv")  # Replace 'image.csv' with the path to your image CSV file
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd

class FashionRecommender:
    def __init__(self, item_features, fashion_data):
        self.item_features = item_features
        self.fashion_data = fashion_data
        self.vectorizer = None
        
    def fit_vectorizer(self, features):
        self.vectorizer = TfidfVectorizer()
        self.item_features = self.vectorizer.fit_transform(features)
        
    def recommend_items(self, input_categories, input_description, n=10):
        if self.vectorizer is None:
            raise ValueError("The vectorizer is not fitted. Please fit the vectorizer before making recommendations.")
        # Combine input categories and description into a single query
        query = ' '.join(input_categories) + ' ' + input_description
        # Transform the query using the vectorizer
        query_features = self.vectorizer.transform([query])
        # Calculate cosine similarity between the query and item features
        similarity_scores = cosine_similarity(query_features, self.item_features)
        # Find indices of the most similar items
        similar_items_indices = similarity_scores.argsort()[0][-n-1:-1][::-1]
        return self.fashion_data.iloc[similar_items_indices]

# Load fashion item dataset
# Assuming df is your existing DataFrame containing fashion item data
# df = pd.read_csv('styles.csv')

# Feature Engineering
df['features'] = df['subCategory'] + ' ' + df['baseColour'] + ' ' + df['articleType'] + ' ' + df['masterCategory'] + ' ' + df['usage'] + ' ' + df['gender'] + ' ' + df['productDisplayName']

# Instantiate the recommender system
recommender = FashionRecommender(None, df)
recommender.fit_vectorizer(df['features'])

# Function to find and display image based on image ID
def find_and_display_image(image_id, image_csv_path):
    try:
        # Load image CSV file
        image_df = pd.read_csv(image_csv_path)
        # Find the image path based on the provided ID
        image_path = image_df[image_df['id'] == image_id]['file_path'].values[0]
        # Open and display the image
        img = Image.open(image_path)
        plt.imshow(img)
        plt.title(f"Budget: {(image_id)/10}")
        plt.axis('off')  # Turn off axis labels
        plt.show()
    except (IndexError, FileNotFoundError):
        # Handle the case where the ID is not found or CSV file is not found
        print(f"Image path not found for ID {image_id}.")

def main():
    input_categories = input("Select Clothing Categories (separated by commas): ").strip().split(',')
    input_description = input("Describe the outfit you desire or the occasion you need an outfit for: ")
    recommendations = recommender.recommend_items(input_categories, input_description)
    if not recommendations.empty:
        print("Recommended Outfits:")
        for index, row in recommendations.iterrows():
            image_id = row['id']  # Assuming 'id' is the column containing unique image IDs
            find_and_display_image(image_id, r"C:\Users\Vaishnavi\Desktop\hackathon copy\image_data.csv")  # Replace 'image.csv' with the path to your image CSV file
    else:
        print("No outfits found for the given criteria.")

if __name__ == "__main__":
    main()
```
